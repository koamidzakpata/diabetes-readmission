```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Affichage amélioré
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 100)
sns.set_theme(style="whitegrid")

print("✅ Bibliothèques importées avec succès !")
```

```python
# Chargement du dataset
df = pd.read_csv('../data/raw/diabetic_data.csv')

# Premier aperçu
print(f"✅ Dataset chargé !")
print(f"📊 Dimensions : {df.shape[0]} lignes × {df.shape[1]} colonnes")
df.head()
```

```python
# Informations générales
print("="*50)
print("INFORMATIONS GÉNÉRALES")
print("="*50)
df.info()

print("\n" + "="*50)
print("STATISTIQUES DESCRIPTIVES")
print("="*50)
df.describe()
```

```python
print("="*50)
print("ANALYSE DES VALEURS MANQUANTES")
print("="*50)

# Les vraies NaN
nan_counts = df.isnull().sum()

# Les '?' cachés dans les colonnes texte
question_mark_counts = (df == '?').sum()

# Tableau récapitulatif
missing = pd.DataFrame({
    'NaN réels': nan_counts,
    'Valeurs "?"': question_mark_counts,
})
missing['Total manquant'] = missing['NaN réels'] + missing['Valeurs "?"']
missing['% manquant'] = (missing['Total manquant'] / len(df) * 100).round(2)

# Afficher uniquement les colonnes avec des manquants
missing_cols = missing[missing['Total manquant'] > 0].sort_values('% manquant', ascending=False)
print(missing_cols)
```

```python
# Visualisation des valeurs manquantes
plt.figure(figsize=(10, 6))
sns.barplot(
    x=missing_cols['% manquant'],
    y=missing_cols.index,
    hue=missing_cols.index,   # ← correction
    palette='Reds_r',
    legend=False               # ← correction
)
plt.title('Pourcentage de valeurs manquantes par colonne', fontsize=14)
plt.xlabel('% manquant')
plt.ylabel('Colonnes')
plt.tight_layout()
plt.savefig('../reports/figures/missing_values.png', dpi=150)
plt.show()
print("✅ Graphique sauvegardé !")
```

```python
print("="*50)
print("DISTRIBUTION PAR ÂGE")
print("="*50)

plt.figure(figsize=(10, 5))
ax = sns.countplot(
    x='age',
    data=df,
    hue='age',
    palette='Blues_d',
    legend=False,
    order=sorted(df['age'].unique())
)

plt.title("Distribution des patients par tranche d'âge", fontsize=14)
plt.xlabel("Tranche d'âge")
plt.ylabel("Nombre de patients")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../reports/figures/age_distribution.png', dpi=150)
plt.show()

# Réadmission par âge
print("\nTaux de réadmission <30 jours par tranche d'âge :")
age_readmit = df[df['readmitted'] == '<30'].groupby('age').size() / df.groupby('age').size() * 100
print(age_readmit.round(2))
```

```python
print("="*50)
print("DURÉE D'HOSPITALISATION vs RÉADMISSION")
print("="*50)

plt.figure(figsize=(10, 5))
sns.boxplot(
    x='readmitted',
    y='time_in_hospital',
    data=df,
    hue='readmitted',
    palette='Blues',
    legend=False,
    order=['NO', '>30', '<30']
)
plt.title("Durée d'hospitalisation selon le statut de réadmission", fontsize=14)
plt.xlabel("Réadmission")
plt.ylabel("Durée (jours)")
plt.tight_layout()
plt.savefig('../reports/figures/time_in_hospital.png', dpi=150)
plt.show()

print("\nMoyenne de durée par statut :")
print(df.groupby('readmitted')['time_in_hospital'].mean().round(2))
```

```python
print("="*50)
print("MÉDICAMENTS & PROCÉDURES vs RÉADMISSION")
print("="*50)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

variables = ['num_medications', 'num_lab_procedures', 'num_procedures']
titres = ['Nombre de médicaments', 'Procédures en labo', 'Procédures médicales']

for ax, var, titre in zip(axes, variables, titres):
    sns.boxplot(
        x='readmitted',
        y=var,
        data=df,
        hue='readmitted',
        palette='Blues',
        legend=False,
        order=['NO', '>30', '<30'],
        ax=ax
    )
    ax.set_title(titre, fontsize=12)
    ax.set_xlabel('Réadmission')
    ax.set_ylabel('')

plt.suptitle('Médicaments & Procédures selon le statut de réadmission', fontsize=14)
plt.tight_layout()
plt.savefig('../reports/figures/medications_procedures.png', dpi=150)
plt.show()

print("\nMoyennes par statut de réadmission :")
print(df.groupby('readmitted')[variables].mean().round(2))
```

```python
print("="*50)
print("INSULINE vs RÉADMISSION")
print("="*50)

plt.figure(figsize=(10, 5))
insulin_readmit = pd.crosstab(
    df['insulin'],
    df['readmitted'],
    normalize='index'
) * 100

insulin_readmit.plot(
    kind='bar',
    figsize=(10, 5),
    colormap='Blues',
    edgecolor='black'
)
plt.title('Statut insuline vs Réadmission (%)', fontsize=14)
plt.xlabel('Statut insuline')
plt.ylabel('Pourcentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Réadmission')
plt.tight_layout()
plt.savefig('../reports/figures/insulin_readmission.png', dpi=150)
plt.show()

print("\nRépartition en % :")
print(insulin_readmit.round(2))
```
